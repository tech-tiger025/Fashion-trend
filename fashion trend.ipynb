##### Analyzing customer purchasing patterns across different fashion categories (e.g., traditional wear, casual, luxury).

```python
# Preview the data
print("Original Data:\n", df.head())
```

```python
# Clean Column Names (Remove extra spaces)
df.columns = df.columns.str.strip()
```

```python
# Check if necessary columns exist
if "product_category" in df.columns and "sales_revenue" in df.columns:
    # Convert 'sales_revenue' column to numeric (in case of text errors)
    df["sales_revenue"] = pd.to_numeric(df["sales_revenue"], errors="coerce")
```

```python
 # Group by Product Category and Sum Total Sales
category_sales = df.groupby("Product_Category")["Sales_Revenue"].sum().reset_index()
```

```python
  # Sort by Sales Revenue in descending order
category_sales = category_sales.sort_values(by="Sales_Revenue", ascending=False)
```

```python
 # Plot the results
plt.figure(figsize=(10, 6))
plt.bar(category_sales["Product_Category"], category_sales["Sales_Revenue"], color=["#4CAF50", "#2196F3", "#FF5722"])
plt.title("Customer Purchasing Patterns Across Fashion Categories")
plt.xlabel("Fashion Category")
plt.ylabel("Total Sales Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
 # Print top-performing categories
print("\nTop Performing Fashion Categories:")
print(category_sales)
```

##### Identifying the impact of seasonality on fashion sales.


```python
# Ensure required columns exist
if "date" in df.columns and "sales_revenue" in df.columns:
    # Convert 'date' column to DateTime format
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
# Ensure the correct column name (Check if 'Date' or something else)
date_column = None
for col in df.columns:
    if "date" in col.lower():  # Look for variations of "date"
        date_column = col
        break
```

```python
if date_column and "sales_revenue" in df.columns:
    print(f"Using '{date_column}' as the date column.")  # Debugging step
```

```python
 # Drop rows where the date conversion failed
df = df.dropna(subset=[date_column])
```

```python
 # Convert 'Date' column to DateTime format
df[date_column] = pd.to_datetime(df[date_column], errors="coerce")
```

```python
 # Check if conversion was successful
if df[date_column].isna().sum() > 0:
        print("Warning: Some date values could not be converted and will be dropped.")
```

```python
 # Drop rows with invalid dates
df = df.dropna(subset=[date_column])
```

```python
# Debug: Check the data type
print(f"Data type of '{date_column}':", df[date_column].dtype)
```

```python
  # Extract Month and Season
df["month"] = df[date_column].dt.month
df["season"] = df["month"].map({
        12: "Winter", 1: "Winter", 2: "Winter",
        3: "Spring", 4: "Spring", 5: "Spring",
        6: "Summer", 7: "Summer", 8: "Summer",
        9: "Autumn", 10: "Autumn", 11: "Autumn"
    })
```

```python
 # Convert 'sales_revenue' to numeric
df["Sales_Revenue"] = pd.to_numeric(df["Sales_Revenue"], errors="coerce")
```

```python
 # Aggregate Sales by Season
season_sales = df.groupby("season")["Sales_Revenue"].sum().reset_index()
```

```python
# Sort Seasons in Logical Order
season_order = ["Winter", "Spring", "Summer", "Autumn"]
season_sales["season"] = pd.Categorical(season_sales["season"], categories=season_order, ordered=True)
season_sales = season_sales.sort_values("season")
```

```python
# Plot Sales by Season
plt.figure(figsize=(8, 5))
plt.bar(season_sales["season"], season_sales["Sales_Revenue"], color=["#4A90E2", "#50E3C2", "#F5A623", "#D0021B"])
plt.title("Seasonal Impact on Fashion Sales")
plt.xlabel("Season")
plt.ylabel("Total Sales Revenue")
plt.show()
```

```python
 # Print Seasonality Insights
print("\nFashion Sales by Season:")
print(season_sales)
```

##### Assessing how pricing affects customer demand.

```python
if "Unit_Price" in df.columns and "Quantity" in df.columns:
    # Convert columns to numeric (handle errors)
    df["Unit_Price"] = pd.to_numeric(df["Unit_Price"], errors="coerce")
    df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
```

```python
 # Drop missing values
df = df.dropna(subset=["Unit_Price", "Quantity"])
```

```python
 # Calculate correlation between price and demand
correlation, p_value = pearsonr(df["Unit_Price"], df["Quantity"])
```

```python
 print(f"Correlation between Price and Demand: {correlation:.2f}")
if p_value < 0.05:
        print("The correlation is statistically significant.")
```

```python
 # Plot Price vs. Demand
plt.figure(figsize=(8, 5))
sns.regplot(x=df["Unit_Price"], y=df["Quantity"], scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.title("Impact of Pricing on Customer Demand")
plt.xlabel("Unit Price")
plt.ylabel("Quantity Sold")
plt.show()
```

##### Compare fashion trends across different regions in England.


```python
if "Region" in df.columns and "Product_Category" in df.columns and "Sales_Revenue" in df.columns:
    # Convert 'sales_revenue' to numeric
    df["Sales_Revenue"] = pd.to_numeric(df["Sales_Revenue"], errors="coerce")
```

```python
 # Group data by region and product category
region_trends = df.groupby(["Region", "Product_Category"])["Sales_Revenue"].sum().reset_index()
```

```python
 # Pivot for visualization
pivot_table = region_trends.pivot(index="Region", columns="Product_Category", values="Sales_Revenue")
```

```python
# Plot the regional fashion trends
pivot_table.plot(kind="bar", figsize=(12, 6), colormap="viridis", alpha=0.8)
plt.title("Fashion Trends Across Regions in England")
plt.xlabel("Region")
plt.ylabel("Total Sales Revenue")
plt.legend(title="Fashion Category", bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
 # Display the data summary
print("\nFashion Sales by Region and Category:")
print(region_trends.sort_values(by="Sales_Revenue", ascending=False))
```
